```python
import pandas as pd
df = pd.read_csv('./CO2 Emissions_Canada.csv')      #데이터 파일 불러오기
df = df.drop(columns = ['Fuel Consumption Comb (mpg)']) #필요없는 컬럼 제거
df
```

```python
print(df.info())   #빈 칼럼 여부 확인하기
```

```python
from sklearn.preprocessing import LabelEncoder #문자 데이터를 수치값으로 변환하기(0~부터 시작 / Make, Vehicle Class, Fuel Type)
le = LabelEncoder()
df['Make'] = le.fit_transform(df['Make'])
df['Vehicle Class'] = le.fit_transform(df['Vehicle Class'])
df['Fuel Type'] = le.fit_transform(df['Fuel Type'])
```

```python
df
```

```python
#정규화
df['Engine Size(L)'] = df['Engine Size(L)']/df['Engine Size(L)'].max()
df['Cylinders'] = df['Cylinders']/df['Cylinders'].max()
df['Fuel Consumption City (L/100 km)'] = df['Fuel Consumption City (L/100 km)']/df['Fuel Consumption City (L/100 km)'].max()
df['Fuel Consumption Hwy (L/100 km)'] = df['Fuel Consumption Hwy (L/100 km)']/df['Fuel Consumption Hwy (L/100 km)'].max()
df['Fuel Consumption Comb (L/100 km)'] = df['Fuel Consumption Comb (L/100 km)']/df['Fuel Consumption Comb (L/100 km)'].max()
df['CO2 Emissions(g/km)'] = df['CO2 Emissions(g/km)']/df['CO2 Emissions(g/km)'].max()
```
